#1. Load the dataset and explore the variables by `pandas`

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
````

# 3. Use the `pandas` to list all the variables/predictors (columns).

````python
# Read the CSV file
data = pd.read_csv('Boston.csv')
# Print column names
print(data.columns)
````

````python
# Create histograms for each column
for col in data.columns:
    plt.hist(data[col])
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {col}')
    plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Boston dataset
data = pd.read_csv("Boston.csv")

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a figure with desired size
fig, ax = plt.subplots(figsize=(10, 7))  # Set your desired width and height

# Create the heatmap using matplotlib
sns.heatmap(
    correlation_matrix, annot=True, cmap="coolwarm", ax=ax  # Pass the axis object
)

# Customize the plot (optional)
plt.title("Correlation Matrix of Boston Housing Data")
plt.tight_layout()  # Adjust spacing automatically

plt.show()
````

# 4. After exploring the dataset, select the `medv` as the response/label 'y' and the rest of the variables as the predictors `X`.

````python
# Assuming you have already loaded the Boston dataset as "data"

# Select the target variable (response)
y = data['medv']

# Select all other columns as predictors (features)
X = data.drop('medv', axis=1)

# Print shapes of X and y to confirm selection
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
````

# 5. Split the dataset into a training set and a testing set. The training set should be 70% of the original dataset and the test set is the rest 30%.
#6. Train a linear regression model on the training set and evaluate its performance on the test set.

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)
mean_squared_error(y_test, y_pred)
````
